```python
from analysis import ResultsAnalysis
import requests
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
plt.ioff()
url = 'https://ai-process-sandy.s3.eu-west-1.amazonaws.com/purge/deviation.json'
resp = requests.get(url=url)
data = resp.json()
```

```python
analysis = ResultsAnalysis(data)
```

```python
paths = analysis.analyze()
print(paths)
```

```python
plt.axis('off')
for path in paths:
    plt.cla()
    plt.clf()
    img = mpimg.imread(path)
    imgplot = plt.imshow(img)
    plt.show()
```
